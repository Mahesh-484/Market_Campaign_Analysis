### Outlier Detection

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

check_cols = ['MntFishProducts','Time_Engaged','Recency', 'MntWines','MntFruits', 'MntGoldProds', 'MntSweetProducts', 'Income', 'MntMeatProducts']

fig, axes = plt.subplots(nrows=len(check_cols) // 2 + len(check_cols) % 2, ncols=2, figsize=(12, len(check_cols) * 2))
axes = axes.flatten()

for i, col in enumerate(check_cols):
    sns.boxplot(y=data[col], ax=axes[i], color='skyblue')
    axes[i].set_title(col, fontsize=10)
    axes[i].set_xlabel('')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#### Violin Plot for Outlier Detection using Z-score

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

check_cols = ['MntFishProducts', 'Recency', 'MntWines','MntFruits', 'MntGoldProds', 'MntSweetProducts', 'Income', 'MntMeatProducts']

z_scores = df[check_cols].apply(zscore)
threshold = 3
outliers = (z_scores > threshold) | (z_scores < -threshold)
outlier_counts = outliers.sum()

fig, axes = plt.subplots(len(check_cols), 1, figsize=(8, len(check_cols) * 2))
for i, col in enumerate(check_cols):
    sns.violinplot(y=df[col], ax=axes[i], color='skyblue')
    axes[i].set_title(col, fontsize=10)
    axes[i].set_xlabel('')

plt.tight_layout()
plt.show()

print("Number of outliers detected per column:")
print(outlier_counts)
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import Normalizer
normal = Normalizer()
data_scaled = normal.fit_transform(data)
optimal_k = 4
pca = PCA(n_components=3)
pca_data = pca.fit_transform(data_scaled)

kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(pca_data)

plt.figure(figsize=(8, 5))
sns.scatterplot(x=pca_data[:, 0], y=pca_data[:, 1], hue=clusters, palette="viridis")
plt.title("K-Means Clustering Visualization with PCA")
plt.show()

print(f"Silhouette Score for K-Means with {optimal_k} clusters: {silhouette_score(pca_data, clusters):.4f}")
```

### DBSCAN Clustering (Finding Epsilon, PCA & Silhouette Score)

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import numpy as np
neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(pca_data)
distances, indices = neighbors_fit.kneighbors(pca_data)

distances = np.sort(distances[:, 4], axis=0)

plt.figure(figsize=(8, 5))
plt.plot(distances)
plt.xlabel("Data Points")
plt.ylabel("Epsilon Distance")
plt.title("K-Distance Graph for Choosing Epsilon")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.02, min_samples=50)
clusters_dbscan = dbscan.fit_predict(pca_data)

plt.figure(figsize=(8, 5))
sns.scatterplot(x=pca_data[:, 0], y=pca_data[:, 1], hue=clusters_dbscan, palette="viridis")
plt.title("DBSCAN Clustering Visualization with PCA")
plt.show()

if len(set(clusters_dbscan)) > 1:
    print(f"Silhouette Score for DBSCAN: {silhouette_score(pca_data, clusters_dbscan):.4f}")
else:
    print("Silhouette Score not available as DBSCAN found a single cluster.")
```
